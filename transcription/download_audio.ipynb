```python
# get list of debates to download
import csv
import os

output_path = './debates/'

with open('list_of_debates.csv', 'r') as file:
    yt_links = {}
    reader = csv.DictReader(file)
    for row in reader:
        yt_links[row['name']] = row['yt_link']

for name, yt_link in yt_links.items():
    # check if the debate has already been downloaded
    if os.path.exists(f'{output_path}/{name}/audio.wav'):
        print(f'{name} already downloaded')
        continue
    ! yt-dlp -x --audio-format wav "{yt_link}" -o "{output_path}/{name}/audio.wav" 

```

# Delete Audio

```python
# delete the audio files when done using
paths = [f'{output_path}/{name}/audio.wav' for name in yt_links.keys()]

for path in paths:
    ! rm "{path}"
```
